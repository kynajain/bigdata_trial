```python
# Importing necessary libraries
import pandas as pd
import spotipy
from spotipy.oauth2 import SpotifyClientCredentials

# Setting up Spotipy with your Spotify API credentials
client_credentials_manager = SpotifyClientCredentials(client_id='CLIENT_ID', client_secret='CLIENT_SECRET_ID')
sp = spotipy.Spotify(client_credentials_manager=client_credentials_manager)

# Reading the CSV file into a DataFrame
df = pd.read_csv("artist_songs.csv")

# Defining a function to search Spotify for the track and get audio features
def get_spotify_audio_features(artist, title):
    query = f"artist:{artist} track:{title}"  # Creating a search query for the artist and title
    results = sp.search(q=query, type='track', limit=1)  #Spotify developer docs and chatgpt 3.5 used for searching for the track on Spotify
    tracks = results['tracks']['items']
    # Getting the track ID of the found track
    track_id = tracks[0]['id']
    # Fetching audio features for the track
    audio_features = sp.audio_features(track_id)[0]
    return audio_features

# Looping through the DataFrame and fetching audio features for each song
for index, row in df.iterrows():
    artist = row['artist']
    title = row['title']
    # Getting audio features for the current track
    features = get_spotify_audio_features(artist, title)
    if features:
        # Printing the audio features if retrieved
        print(f"Audio features for {title} by {artist}: {features}")
    else:
        # Printing a message if features could not be retrieved for debugging
        print(f"Failed to retrieve features for {title} by {artist}")
```

```python
# Initializing a dictionary to hold cumulative valence scores and track counts
artist_valence = {}

# Defining a function to fetch a track's audio features from Spotify
def get_audio_features(artist, title):
    # Searching for the track on Spotify
    results = sp.search(q=f"artist:{artist} track:{title}", type='track', limit=1)
    tracks = results['tracks']['items']
    if tracks:
        # Getting the track ID of the found track
        track_id = tracks[0]['id']
        # Returning the audio features for the track
        return sp.audio_features(track_id)[0]
    else:
        # Returning None if no track is found
        return None

# Looping through each song in the DataFrame, fetching audio features, and calculating total valence
for index, row in df.iterrows():
    artist = row['artist']
    title = row['title']
    features = get_audio_features(artist, title)
    if features:
        # Checking if the artist is already in the dictionary
        if artist not in artist_valence:
            artist_valence[artist] = {'total_valence': 0, 'count': 0}
        # Updating the total valence and count for the artist
        artist_valence[artist]['total_valence'] += features['valence']
        artist_valence[artist]['count'] += 1

# Creating a DataFrame for average valence scores
valence_df = pd.DataFrame(columns=['Artist', 'Average Valence'])

# Looping through the dictionary to calculate the average valence for each artist
for artist, data in artist_valence.items():
    # Calculating the average valence
    average = data['total_valence'] / data['count']
    valence_df.loc[len(valence_df)] = {'Artist': artist, 'Average Valence': average}

# Saving the DataFrame to a csv file
valence_df.to_csv('average_valence_scores.csv', index=False)
print("Saving average valence scores to 'average_valence_scores.csv'.")
```
